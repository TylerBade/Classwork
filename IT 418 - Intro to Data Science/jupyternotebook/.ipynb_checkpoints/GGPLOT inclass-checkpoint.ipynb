```python
ggplot(diamonds, aes(x='price', color='clarity')) + geom_density()
```

```python
ggplot(aes(x='carat',y='price'), data=diamonds) + geom_point()
```

```python
ggplot(aes(x='carat',y='price', color='clarity'), data=diamonds) + geom_point()
```

```python
ggplot(aes(x='carat',y='price', color='clarity'), data=diamonds) + \
geom_point() + scale_color_brewer(type='seq',palette='Reds')
```

```python
ggplot(aes(x='carat',y='price', color='clarity'), data=diamonds) + \
geom_point() + scale_color_brewer(type='div',palette=2)
```

```python
ggplot(aes(x='carat',y='price', color='clarity'), data=diamonds) + \
geom_point() + scale_color_brewer(type='qual')
```

```python
ggplot(mpg, aes(x='cty', y='hwy')) + geom_point()
```

```python
ggplot(mpg, aes(x='cty', y='hwy')) + geom_point() + \
ggtitle("City vs. Highway Miles per Galon")
```

```python
ggplot(mpg, aes(x='cty', y='hwy')) + geom_point() + \
ggtitle("City vs. Highway Miles per Galon") + \
xlab("City MPG(Miles per Gallon)") + ylab("Highway MPG(Miles per Gallon)")
```

```python
ggplot(mpg, aes(x='cty', y='hwy', color='class')) + geom_point() + \
ggtitle("City vs. Highway Miles per Galon") + \
xlab("City MPG(Miles per Gallon)") + ylab("Highway MPG(Miles per Gallon)") +\
scale_color_brewer(type='qual')
```

```python
ggplot(mpg, aes(x='cty', y='hwy', color='class')) + geom_point() + \
ggtitle("City vs. Highway Miles per Galon") + \
xlab("City MPG(Miles per Gallon)") + ylab("Highway MPG(Miles per Gallon)") +\
scale_color_brewer(type='qual') + facet_wrap(x='year', y='cyl')
```
